```python
# mutable object - list, set
a = [1, 2, 3]
print("id(a)= ", id(a))
a[0] = 5
print('id(a)=", id(a))

# immutable object - str, int, float
s = 'apple'
print('id(s)= ', id(s))
print(s[0])
#s[0] = 'b'

# how to make a shallow copy? - copy.copy() or slicing copy
# shallow copy of mutable object
a = [1, 2, 3]
b = a # shallow copy
b[0] = 5
print('a=', a)
print('b=', b)

print('id(a) = ', id(a))
print('id(b) = ', id(b))

# shallow copy of immutable object
s = 'apple'
t = s
print('s=', s)
print('t=', t)
print('id(s)', s)
print('id(t)', t)

# what if new value is assigned to t?
t = 'banana'
print('id(s)', s)
print('id(t)', t)

# shallow copy of mutable object - list
a = [1, 2, 3]
b = a[:] # list slicing is a kind of shallow copy
print('id(a)', id(a))
print('id(b)', id(b))

print(a == b)
print(a is b)

b[0] = 5
print(a)
print(b)

# shallow copy of nested list
a = [[1, 2], [3, 4]]
b = a[:]

print(id(a))
print(id(b))
print(id(a[0])
print(id(b[0])

a[0] = [8,9]
print(a)
print(b)

print(id(a[0]))
print(id(b[0]))

print(id(a[1]))
print(id(b[1]))

a[1].append(5)
print(a)
print(b)



# deepcopy 
import copy
a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)

a[1].append(5)
print(a)
print(b)
```
